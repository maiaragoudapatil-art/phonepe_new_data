```python
import pandas as pd
import json
import os

# Re-defining base_path and loading data_list and df to ensure df is available
base_path = "/content/pulse_data/content/pulse_data/pulse-master/data/aggregated/transaction/country/india/state"
states = os.listdir(base_path)

data_list = []

for state in states:
    state_path = os.path.join(base_path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, 'r') as f:
                content = json.load(f)

                # Skip if no data
                if content['data']['transactionData'] is None:
                    continue

                for item in content['data']['transactionData']:
                    data_list.append({
                        "State": state,
                        "Year": int(year),
                        "Quarter": int(file.strip(".json")),
                        "Transaction_Type": item['name'],
                        "Count": item['paymentInstruments'][0]['count'],
                        "Amount": item['paymentInstruments'][0]['amount']
                    })

df = pd.DataFrame(data_list)

# Ensure column names are consistent before aggregation
df.columns = df.columns.str.lower().str.replace(" ", "_")

aggregated_transaction = df.groupby('year')['amount'].sum().reset_index()
aggregated_transaction = aggregated_transaction.rename(columns={'amount': 'total_amount'})
aggregated_transaction = aggregated_transaction.sort_values(by='year')
print(aggregated_transaction)
```

```python
df.info()
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
df['year'] = df['year'].astype(int)
df['quarter'] = df['quarter'].astype(int)
df['count'] = df['count'].astype(int)
df['amount'] = df['amount'].astype(float)
```

```python
df['state'] = df['state'].str.replace("-", " ").str.title()
#andhra-pradesh → Andhra Pradesh
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['avg_transaction_value'] = df['amount'] / df['count']
```

```python
threshold = df["amount"].mean() * 5

outliers = df[df["amount"] > threshold]

outliers.head()
```

```python
volatility = (
    df.groupby("state")["amount"]
    .std()
    .sort_values(ascending=False)
)

volatility.head(10)
```

```python
consistency = df.groupby("state")["amount"].agg(["mean", "std"])
consistency["score"] = consistency["mean"] / consistency["std"]

consistency.sort_values("score", ascending=False).head(10)
```

```python
efficiency = df.groupby("state").agg({
    "amount": "sum",
    "count": "sum"
})

efficiency["efficiency"] = efficiency["amount"] / efficiency["count"]

efficiency.sort_values("efficiency", ascending=False).head(10)
```

```python
rank_df = (
    df.groupby(["year", "state"])["amount"]
    .sum()
    .reset_index()
)

rank_df["rank"] = rank_df.groupby("year")["amount"].rank(ascending=False)

rank_df.head()
```

```python
category_growth = (
    df.groupby(["transaction_type", "year"])["amount"]
    .sum()
    .unstack()
)

category_growth.pct_change(axis=1) * 100
```

```python
state_total = df.groupby("state")["amount"].sum()
share = state_total / state_total.sum()

top_5_share = share.sort_values(ascending=False).head(5).sum()

top_5_share
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_states = df.groupby('state')['amount'].sum().reset_index()
top_states = top_states.rename(columns={'amount': 'total_amount'})
top_states = top_states.sort_values(by='total_amount', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_states, x='state', y='total_amount')
plt.xticks(rotation=45)
plt.title("Top 10 States by Transaction Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

txn_type = df.groupby('transaction_type')['amount'].sum().reset_index()
txn_type = txn_type.rename(columns={'amount': 'total_amount'})
txn_type = txn_type.sort_values(by='total_amount', ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(data=txn_type, x='transaction_type', y='total_amount')
plt.xticks(rotation=45)
plt.title("Transaction Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(txn_type['total_amount'],
        labels=txn_type['transaction_type'],
        autopct='%1.1f%%')
plt.title("Transaction Share by Category")
plt.show()
```
